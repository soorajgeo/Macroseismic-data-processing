```python
import pandas as pd
```

```python
df = pd.read_csv('values.csv')
df.columns
```

```python

# Define custom mappings for each specified column
custom_mappings = {
    'Situation': dict(zip(df['Situation'], df['Value1'])),
    'Your feeling': dict(zip(df['Your feeling'], df['Value2'])),
    'Others feeling': dict(zip(df['Others feeling'], df['Value3'])),
    'Duration': dict(zip(df['Duration'], df['Value4'])),
    'Noise': dict(zip(df['Noise'], df['Value5'])),
    'Awake': dict(zip(df['Awake'], df['Value6'])),
    'Reaction': dict(zip(df['Reaction'], df['Value7'])),
    'Response': dict(zip(df['Response'], df['Value8'])),
    'Building type': dict(zip(df['Building type'], df['Value9'])),
    'Swinging': dict(zip(df['Swinging'], df['Value10'])),
    'Shelves': dict(zip(df['Shelves'], df['Value11'])),
    'Picture': dict(zip(df['Picture'], df['Value12'])),
    'Furniture': dict(zip(df['Furniture'], df['Value13'])),
    'Heavy objects': dict(zip(df['Heavy objects'], df['Value14'])),
    'House damage': dict(zip(df['House damage'], df['Value15'])),
    
}

# Create the nested dictionary based on the custom mappings
nested_dict = {col: custom_mappings[col] for col in custom_mappings}

print(nested_dict)

```

```python
df2 = pd.read_csv('observations.csv')
df2.drop(columns='Person',inplace=True)
df2.head()
```

```python
# Apply label encoding with custom values
for col, values in nested_dict.items():
    df2[col] = df2[col].map(values)

df2
```

```python
communalities = {'Situation':0.04940218, 'Your feeling':0.15383976, 'Others feeling':0.11502347, 'Duration':0.12901107, 'Noise':0.0975182,
       'Awake':0.00618498, 'Reaction':0.150554, 'Response':0.12187119, 'Building type':0.08150183, 'Swinging':0.09555509, 'Shelves':0.11973431,
       'Picture':0.09833884, 'Furniture':0.13933999, 'Heavy objects':0.05697579, 'House damage':0.08514933}
```

```python
df_communality = df2.apply(lambda col: col * communalities[col.name])
df_communality
```

```python
df_communality['MSKWI'] = 0.916 * df_communality.sum(axis=1)
df_communality
```

```python
df_communality['MSK64'] = (0.96 * df_communality['MSKWI'])+0.1
df_communality
```

```python
df_communality['MSK64'] = df_communality['MSK64'].round()
df_communality
```
